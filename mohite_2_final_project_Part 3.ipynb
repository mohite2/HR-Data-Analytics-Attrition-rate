## Ploting line plot for montly Income:

```python
working_hour = att_hr_data.groupby('percentsalaryhike')['monthlyincome'].mean().reset_index()

# Getting a look at the values of the fields in the specified columns

attr_income = np.array(working_hour['monthlyincome'].astype('float'))
attr_work_years = np.array(working_hour['percentsalaryhike'].astype('float'))

attr_income
attr_work_years
```

```python
# Defining the scales:

x_scale = bqplot.LinearScale(min=10,max=30) 
y_scale = bqplot.LinearScale(min=0, max=40000)
width_line = bqplot.LinearScale()

#Defining the axis 
axis_xcl = bqplot.Axis(label='Percentage of hike', scale=x_scale)
axis_ycl = bqplot.Axis(label='Mean Monthly Income', scale=y_scale, 
                    orientation='vertical', side='left')

#Plotting the line graph
lineplt = bqplot.Lines(x = attr_work_years,
                               y = attr_income, 
                              scales={'x':x_scale, 'y':y_scale,'width': width_line,},
                              display_legend=True, 
                              marker_size=20, marker = 'circle', colors = ['DarkOrange'], 
                              opacities = [0.7], line_style = 'solid',stroke_width = 3
                                      )


Lineplot = bqplot.Figure(marks = [lineplt], axes = [axis_xcl, axis_ycl],
                         title='Hike % vs Monthly Income', legend_location='top-left')
Lineplot
```

From the above plot, we can see that monthly income decreases as the pecentage of hike increases.
```python
## https://www.geeksforgeeks.org/python-pandas-series-xs/ --- reference
##https://stackoverflow.com/questions/17921010/how-to-query-multiindex-index-columns-values-in-pandas

att_hr_data_2 = attr_pvt.iloc[:, attr_pvt.columns.get_level_values('age')=='jobrole'] ## get the values 

jobroles_data =att_hr_data_2.reset_index().xs('jobrole', axis=1, drop_level=True) ## converting series to dataframe
jobroles_data = jobroles_data.reset_index()

job_role_title = jobroles_data['jobrole'].values.tolist()
jobroles_index = jobroles_data['index'].values.tolist()


### https://stackoverflow.com/questions/18012505/python-pandas-dataframe-columns-convert-to-dict-key-and-value

job_roles_dict = dict(zip(jobroles_index, job_role_title)) ##store all job roles vlaues in dictnary
job_roles_dict

##attr_data2 = att_hr_data.set_index(["jobrole"])
```

```python
working_years_data = att_hr_data.groupby(['jobrole','percentsalaryhike'])['monthlyincome'].mean().reset_index().reset_index()
working_years_data

```

## Creating Interactivity between Heatmap and Line plot

```python
# prep_notebook_week07_part1 ---- reference 

mySelectedLabel = ipywidgets.Label()


def get_data_value2(change):
    i,j = change['owner'].selected[0]
    v = attr_pvt_values[i,j] # grab data value
    clicked_job = job_roles_dict[i]
    mySelectedLabel.value = 'Selected Job_Roles - '+str(clicked_job)+', Mean Montly Income = ' + str(v) # set our label
    working_hour = working_years_data[working_years_data['jobrole']==clicked_job]
    
    lineplt.x = np.array(working_hour['percentsalaryhike']).astype('float') # Changing X-Axis for 
    lineplt.y = np.array(working_hour['monthlyincome']).astype('float')
    axis_xcl.label='Percentage Hike for - ' + str(clicked_job)
    axis_ycl.label='Mean Monthly Income for - ' + str(clicked_job)
    x_scale.min = lineplt.x.min()
    x_scale.max = lineplt.x.max()
    y_scale.min = lineplt.y.min()
    y_scale.max = lineplt.y.max()
   # states = {'1':}
   


heat_map.observe(get_data_value2, 'selected')

#Dashboarding
Gridheatmap.layout.min_width='500px'
Lineplot.layout.min_width='500px'


plots=ipywidgets.HBox([Gridheatmap,Lineplot])

ipywidgets.VBox([mySelectedLabel,plots])
```
