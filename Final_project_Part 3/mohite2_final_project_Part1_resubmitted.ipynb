### here we can see the heatmap for attrition rate with no employee left jon by job role and age

```python

# Scales for color, x-axis and y-axis
col_sc = bqplot.ColorScale(scheme = 'RdPu',min=np.nanmin(attr_pvt), 
                           max=np.nanmax(attr_pvt))
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()

# Axis for color, x-axis and y-axis
c_ax = bqplot.ColorAxis(scale = col_sc, 
                        orientation = 'vertical', 
                        side = 'right')
x_ax = bqplot.Axis(scale = x_sc,label='Employees Age', label_color = "RoyalBlue",x_offset = "1.5px")
y_ax = bqplot.Axis(scale = y_sc, label='Job Role of Employees',label_color = "RoyalBlue",
                   orientation = 'vertical')

# Marks for color, x-axis and y-axis
heat_map = bqplot.GridHeatMap(color = attr_pvt,
                              scales = {'color': col_sc,
                                        'row': y_sc,
                                        'column':x_sc },
                              interactions = {'click': 'select'}, # make interactive on click of each box
                              anchor_style = {'fill':'blue'}, # to make our selection blue
                              selected_style = {'opacity': 1.0}, # make 100% opaque if box is selected
                              unselected_style = {'opacity': 0.8}) # make a little see-through if not

# Plotting the heatmap
Gridheatmap = bqplot.Figure(marks = [heat_map], 
                    axes = [c_ax, y_ax, x_ax])
Gridheatmap
```

### Ploting line plot for montly Income:

```python
attr_data2 = att_hr_data.groupby('totalworkingyears')['monthlyincome'].mean().reset_index()

# Getting a look at the values of the fields in the specified columns

attr_income = np.array(attr_data2['monthlyincome'].astype('float'))
attr_work_years = np.array(attr_data2['totalworkingyears'].astype('float'))

attr_income
attr_work_years
```

```python
# Defining the scales:

x_scale = bqplot.LinearScale(min=0,max=40) 
y_scale = bqplot.LinearScale(min=0, max=20000)
width_line = bqplot.LinearScale()

#Defining the axis 
axis_xcl = bqplot.Axis(label='Total Working years  of Employee', scale=x_scale)
axis_ycl = bqplot.Axis(label='Mean Monthly Income', scale=y_scale, 
                    orientation='vertical', side='left')

#Plotting the line graph
lineplt = bqplot.Lines(x = attr_work_years,
                               y = attr_income, 
                              scales={'x':x_scale, 'y':y_scale,'width': width_line,},stroke_width=3,
                              display_legend=True
                                      )


Lineplot = bqplot.Figure(marks = [lineplt], axes = [axis_xcl, axis_ycl],
                         title='Total work years vs Monthly Income as indicator of reduced attrition', legend_location='top-left')
Lineplot
```

### What is the "name" of the dataset?

Attrition analysis of HR data: 

It is often very costly to hire new employee in place of the current by differetial value over 200% of the annual salary as quoted by the AmericanProgress.org.The new hiring process, the employement training, on-boarding bonus are the extra charges when company initiates the new hiring process after the talented employee quites the job.The various factors such as lost prodcitivity, low work-life balance, job role engagement, education, recruitment and total wages are some of the factor that affects the attrition rate of the employee.Losing an employee that’s in the top 1% of performers could mean the difference between growth and decline. Hence, HR analytics evelove around decresing the attrition rate while increasing the employee enagement by monitoring the realtive factors at large.Data analysis can be breakdown of distance from home by job role and attrition’ or ‘compare average monthly income by education and attrition. HR analytics can help improve retention through a churn analysis that looks at data points like:

Attrition by department Attrition by estimated commute time Similar attributes of employees with longer emlopyment Similar attributes of employees who leave within 1 year Onboarding experience Survey data Qualitative data such as employee interviews Employee performance data to forecast future attrition

Through this data-driven approach, HR analytics can illuminate the major causes of attrition, and new policies, along with training programs, can be put in place to help mitigate the problem. For example, data might show that high-aspiration employees are not challenged or employees are frustrated with a certain management style using jobstaisfaction. Human resources analysis will reveal these issues, and then it will be up to leadership to act.`
